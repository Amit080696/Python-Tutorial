```python
import pandas as pd

# Load the JSON data from a file
json_data = {
    "employees": [
        {
            "name": "John",
            "age": 30,
            "address": {
                "city": "New York",
                "street": "123 Main St"
            },
            "skills": ["Python", "SQL"]
        },
        {
            "name": "Jane",
            "age": 28,
            "address": {
                "city": "Los Angeles",
                "street": "456 Elm St"
            },
            "skills": ["Java", "C#"]
        }
    ]
}

# Normalize the JSON data
df = pd.json_normalize(json_data, record_path='employees')

# Display the DataFrame
print(df)
```
